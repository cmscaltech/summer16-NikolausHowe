```python
import xgboost as xgb
class XGBoostClassifier:
    def __init__( self , params ,num_round=10, val=0.8):
        self.params = params
        self.num_round=num_round
        self.validation = val
    def fit(self, X,y):
        n_train = int(X.shape[0]*self.validation)
        dtrain = xgb.DMatrix( X[:n_train], label = y[:n_train])
        dtest =  xgb.DMatrix( X[n_train:], label = y[n_train:])
        evallist  = [(dtest,'eval'), (dtrain,'train')]

        self.model = xgb.train( self.params, dtrain, self.num_round, evallist )

    def predict_proba(self, X):
        xgmat = xgb.DMatrix(X)
        ## fake the output
        res= np.zeros ((len(X),2))
        pred = self.model.predict(xgmat)
        res[:,1] = pred
        return res
    
```

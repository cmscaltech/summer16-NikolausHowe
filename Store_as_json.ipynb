## Store as json

```python
import json
```

### Write to json

```python
# Initialize the json dictionary with empty list of particles
particles = { 'particles' : [] }

# How far we want to start in each event (set between 0 and 40,000)
my_seed = 50

# Prepare the data

# Gammas
for i in range(10):
    event = signal[i+my_seed]
    event_dict = {
        'event_id' : i,
        'pid' : 'gamma',
        'Ex' : 100,
        'Ey' : 0,
        'Ez' : 0,
        'EArray' : event.tolist()
    }
    particles['particles'].append(event_dict)
    
# Pi0s
for i in range(10):
    event = signal[i+my_seed]
    event_dict = {
        'event_id' : i+10,
        'pid' : 'pi0',
        'Ex' : 100,
        'Ey' : 0,
        'Ez' : 0,
        'EArray' : event.tolist()
    }
    particles['particles'].append(event_dict)
    
# Write JSON data to file
with open('particles.json', 'w') as outfile:
     json.dump(particles, outfile)
```

### Read from json

```python
# Read the JSON file
with open('particles.json', 'r') as infile:
     data = json.load(infile)
```
